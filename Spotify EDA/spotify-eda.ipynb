## Representing the null values in a graph

```python
#Percentage of null values
null_vals = (spotify.isnull().sum() / len(spotify) *100)

#plotting a bar graph
plt.figure(figsize = (10,6))
null_vals.plot(kind = 'bar')
plt.title("Percentage of Null values by columns", fontsize = 25)
plt.xlabel("Columns", fontsize=20)
plt.ylabel("Percentage of Null Values", fontsize=20)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
